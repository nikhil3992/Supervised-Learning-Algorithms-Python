```python
import csv
import numpy as np
from sklearn import tree
import math


#Decision Tree Digit Recognition without any pruning

# Opening the opt_digits.raining.csv file
with open('optdigits_raining.csv') as trainingFile:
    reader = csv.reader(trainingFile)
    
    # Declaring variables to store the data
    # X is for storing the attributes and Y is for storing the labels
    X = []
    Y = []
    
    # Reading each line from the file and appending the data to X and Y
    for row in reader:
        X.append(row[:64])
        Y.append(row[64])
    
    # num_correct variable to measure accuracy
    num_correct = 0;

    # classifying the data using decision tree classifier
    clf = tree.DecisionTreeClassifier();
    clf = clf.fit(X, Y)
    
    # Measuring the accuracy of the classifier for training dataset
    for i in range(0,len(X)):
            output_predicted = clf.predict([X[i]])
            originalOutput = Y[i]
        
            # if difference is zero, then the classifier has correctly predicted the output
            difference = int(output_predicted[0]) - int(originalOutput)
            if(np.absolute(difference) == 0):
                num_correct = num_correct + 1;
    
    
    # printing the accuracy of training dataset
    accuracy_Training = (num_correct / len(X)) * 100
    print("Accuracy percentage of Training Dataset without pruning is ")
    print(accuracy_Training);
    
    
    
# Using the classifier on the testing dataset

# Opening the optdigits_test.csv file
with open('optdigits_test.csv') as testingFile:
    testReader = csv.reader(testingFile)
    
    # Declaring variables to store the data
    # X_test is for storing the test dataset attributes and Y_test is for storing the test dataset labels
    X_test = []
    Y_test = []
    
    #Reading each line from the file and appending the data to X_test and Y_test
    for row in testReader:
        X_test.append(row[:64])
        Y_test.append(row[64])

    num_correct = 0;
    length = len(X_test)
    
    # Measuring the accuracy of the classifier for testing dataset
    for i in range(0,length):
        output_predicted = clf.predict([X_test[i]])
        originalOutput = Y_test[i]        
        # if difference is zero, then the classifier has correctly predicted the output
        difference = int(output_predicted[0]) - int(originalOutput)
        if(np.absolute(difference) == 0):
            num_correct = num_correct + 1;
    
    accuracy = (num_correct /len(X_test)) * 100
    print("Accuracy percentage of Testing Dataset without pruning is ")
    print(accuracy);
```
